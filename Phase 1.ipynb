```
!pip install clean-text
!pip install unidecode
import numpy as np
import pandas as pd
```

```
#Reading the language converted dataset
df = pd.read_excel('Text_Translate_File_Main.xlsx')
df.drop(columns = ['Unnamed: 0'], inplace = True)
df.head()
```

```
#Removing errors in the release date columne and the reviews column.
df['review_score_desc'].replace(' user reviews','', regex = True, inplace = True)
df['review_score_desc'].replace('No',0, regex = True, inplace = True)
df['review_score_desc'].unique()
```

### Removing the Emoji's

```
from cleantext import clean

#Cleaning using clean-text library
df['name'] = df['name'].apply(lambda x: clean(x, no_emoji=True))
df['name_translated'] = df['name_translated'].apply(lambda x: clean(x, no_emoji=True))
```

```
df
```

### Cleaning null values

```
df.isna().sum()
```

```
df
```

```
df.loc[df['categories'].isna() & df['developers'].isna() & df['publishers'].isna() & df['genres'].isna()]
```

Wherever the developers, publishers, categories and genre are missing, it indicates that the game was either cancelled or that it was a demo or playtest that was released. Hence we can safely drop all these rows.

```
df.dropna(inplace = True)
df.isna().sum()
```

```
#Converted the release dates to Datetime format
df['release_date'].replace('Not Released',np.nan,inplace = True)
df['release_date'].astype('datetime64[ns]')
df.head()
```
